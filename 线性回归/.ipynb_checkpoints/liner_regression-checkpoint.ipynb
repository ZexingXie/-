## 线性回归



n个训练数据，其中数据的自变量维度为d，因变量维度为1,进行回归分析

样本：$$(x^{(1)}_1,x^{(1)}_2,...,x^{(1)}_d,y^{(1)}) , ... ,(x^{(n)}_1,x^{(n)}_2,...,x^{(n)}_d,y^{(n)})$$


    
   首先假设问题中自变量和因变量可以用线性关系拟合，可以构造预测函数
   
    

即于任意一个样本$x^{?}=(x_1^{?},x_2^{?},...,x_d^{?})$，有估计函数
    $$g(x)=\sum_{i=0}^{d}\theta_i*x_i$$

选择最小二乘法来衡量所有m个样本误差$$J(\theta) = \frac{1}{2}\sum_{j=1}^{m}(\sum_{i=0}^{d}\theta_i*x^{(j)}_i-y^{(j)})^2$$
    

```python
import numpy as np
```

```python

class LinerRegression(object):
    def __init__(self,learning_rate=0.01, max_iter=100,seed=None):
        np.random.seed(seed)
        self.lr = learning_rate
        self.max_iter = max_iter
        self.loss_arr = []
        
    def fit(self,x,y):
        # fit 要求 x的维度>1, y的维度=1
        self.x = np.array(x)
        self.y_true = np.array(y)
        if x.shape[1] is None or x.shape[1]==1:
            self.w = np.random.normal(1,0.1)
        else:
            self.w = np.random.normal(size=(x.shape[1],1)).reshape(x.shape[1],1)
        self.b = np.random.normal(1, 0.1)
        for i in range(self.max_iter):
            self._train_step()
            self.loss_arr.append(self.loss())
        print('finish train %d'%self.max_iter)
        
    def _f(self,x,w,b):
        return np.dot(x, w)+b
    
    def predict(self,x=None):
        if x is None:
            x = self.x
        y_pred = self._f(x,self.w,self.b)
        return y_pred
    
    def _train_step(self):
        d_w, d_b = self._calc_gradient()
        self.w = self.w - self.lr * d_w
        self.b = self.b - self.lr * d_b
        return self.w, self.b
    
    def loss(self,y_true=None, y_pred=None):
        #if y_true == None or y_pred == None:  注意这里不要用==None, ndarray会 ambiguous
        if y_true is None or y_pred is None:
            y_pred = self.predict()
            y_true = self.y_true
        return np.mean((y_pred - y_true)**2)
        
        
    # 使用batch梯度下降方法 _calc_gradient 计算梯度方向
    def _calc_gradient(self):
        d_w = np.mean(self.x * (- self.y_true + self.predict()),axis=0).reshape(self.x.shape[1],1)
        d_b = np.mean((- self.y_true + self.predict()),axis=0)
        return d_w,d_b

```

使用batch梯度下降方法 _calc_gradient 更新learning参数。原理上是对每个参数沿着梯度方向下降，利用样本x来估计参数


$$\theta_i =\theta_i - \alpha \frac{\partial J(\theta)}{\partial \theta_i}=\theta_i - \alpha*\sum_{j=1}^{m}(\sum_{i=0}^{d}\theta_i*x^{(j)}_i-y^{(j)})*x_i $$


简化表示为：$$\theta_i =\theta_i - \alpha*\sum_{j=1}^{m}(g(x^{j})-y^{(j)})*x_i $$


## 测试

```python
# -*- coding: utf-8 -*-
# import Ipynb_importer
from mpl_toolkits.mplot3d import Axes3D  # 空间三维画图
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data generation
np.random.seed(22)
# data_size = 100
data_size = [100,2]
x = np.random.uniform(low=1.0, high=10.0, size=data_size)
w = np.array([20,11]).reshape(x.shape[1],1)
y = np.dot(x, w)

if x.shape[1] == 2:
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(x[:,0],x[:,1], y)
    # 添加坐标轴(顺序是Z, Y, X)
    ax.set_zlabel('Z', fontdict={'size': 15, 'color': 'red'})
    ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
    ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
```

```python
# train the liner regression model
regr = LinerRegression(learning_rate=0.01, max_iter=1, seed=314)
regr.fit(x, y)
_y = regr.predict()


if x.shape[1] == 2:
    fig = plt.figure()
    ax = Axes3D(fig)
    # ax.scatter(x[:,0],x[:,1], y)
    # 添加坐标轴(顺序是Z, Y, X)
    ax.set_zlabel('Z', fontdict={'size': 15, 'color': 'red'})
    ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
    ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
    ax.scatter(x[:,0],x[:,1], _y)
    plt.show()
```

```python
class LinerRegression_L2(LinerRegression):
    def __init__(self,learning_rate=0.01, max_iter=100,seed=None):
        LinerRegression.__init__(self,learning_rate=learning_rate, max_iter=max_iter,seed=seed)
        
    def _calc_gradient(self):
        d_w = np.mean(self.x * (- self.y_true + self.predict()),axis=0).reshape(self.x.shape[1],1)+2*self.w
        d_b = np.mean((- self.y_true + self.predict()),axis=0)+2*self.b
        return d_w,d_b

```

L2正则化，简单来看就是在$J(\theta)$后增加对$\theta$的约束

$$J(\theta) = J(\theta)+\sum_{i=0}^d{ \theta_i^2}$$
所以求导多增加$2\theta$

```python
# train the liner regression model
regr2 = LinerRegression_L2(learning_rate=0.01, max_iter=10, seed=314)
regr2.fit(x, y)
regr2.loss_arr
```

```python
# train the liner regression model
regr = LinerRegression(learning_rate=0.01, max_iter=10, seed=314)
regr.fit(x, y)
_yl0 = regr.predict()
regr.loss_arr
```

L2正则化使得简单问题难以收敛,但当数据量不足训练高维度数据L2表现如下具体的优势
，为此构造维度更高例子

```python
# train / test split
data_size = [100,100]
x = np.random.uniform(low=1.0, high=10.0, size=data_size)
w = np.random.uniform(size=(data_size[1],1))
y = np.dot(x, w)
shuffled_index = np.random.permutation(data_size[0])
x = x[shuffled_index]
y = y[shuffled_index]
split_index = int(data_size[0] * 0.7)
x_train = x[:split_index]
y_train = y[:split_index]
x_test = x[split_index:]
y_test = y[split_index:]
```

```python
regr = LinerRegression(learning_rate=0.0001, max_iter=5000, seed=314)
regr2 = LinerRegression_L2(learning_rate=0.0001, max_iter=5000, seed=314)
regr.fit(x_train,y_train)
regr2.fit(x_train,y_train)
print(regr.loss_arr[-1])
print(regr2.loss_arr[-1])
```

```python
y_test_l0 = regr.predict(x_test)
print(regr.loss(y_test,y_test_l0))
y_test_l2 = regr2.predict(x_test)
print(regr.loss(y_test,y_test_l2))
```

表明了L2提高了模型对规律数据的收敛能力，并提高泛化力。总结为什么L2有效，从查阅的资料个人理解为：
   

 上诉构造的$g(x)$与真实y的误差符合高斯分布，根据极大似然

$$L(\theta)=\prod_{i=1}^{m}\frac{1}{\sqrt{2\pi\mathrm{Cov}}  }*exp(\frac{(y^i-g(x^i))^2}{2\mathrm{Cov}})=m*\log{\frac{1}{\sqrt{2\pi\mathrm{Cov}}}}-\frac{1}{\mathrm{Cov}}*g(x)$$

当Cov越大整体的极大似然越大，导致g(x)缺少泛化力只针对训练样本拟合好。反之Cov小更能约束g(x)。更更严密的解释是从Bayes学派https://lumingdong.cn/detailed-summary-of-regularization.html
